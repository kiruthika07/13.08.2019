```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Normal Distrivution
values= np.random.normal(150,25, 1000)
print('Normal Distribution')
plt.hist(values,100,density=True)
plt.show()
```

```python
#Uniform Distribution
values_uniform= np.random.uniform(-10,10,5000)
count, bins, ignored = plt.hist(values_uniform, 60,density=True)
print('Uniform Distribution')
plt.show()
```

```python
#Poisson Distribution
values_pois = np.random.poisson(10.0, 50000)
plt.hist(values_pois, 10,density=True)
print('Poisson Distribution')
plt.show()
```

```python
#Binomial Distribution
n=1000
p=0.255
values_binom = np.random.binomial(n, p, 10000)
print('Binomial Distribution')
plt.hist(values_binom,20,density=True)
plt.show()
```

```python
print('Exponential Distribution')
target = 500
beta = 1.0/target
values_expo = np.random.exponential(beta, 25000)
plt.hist(values_expo, bins=250,lw=0,alpha=.8)
plt.plot([0,max(values_expo)],[target,target],'r-')
plt.ylim(0,target*1.1)
plt.show()
```

```python
print('Random Distribution')
values_ran=np.random.choice(range(1000,5000),10)
plt.hist(values_ran,5)
plt.show()
```
